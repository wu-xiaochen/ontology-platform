## 2. 导入核心模块

```python
from ontology.reasoner import OntologyReasoner, ConfidenceLevel
from ontology.auto_learn import AutoLearnEngine
from confidence import ConfidenceCalculator
```

## 3. 创建Agent知识库

```python
# 初始化推理引擎
reasoner = OntologyReasoner()

# 添加供应商数据
suppliers = [
    {"id": "SUP001", "name": "Acme Components", "on_time_rate": 0.91, "quality_score": 0.88},
    {"id": "SUP002", "name": "Global Parts Ltd", "on_time_rate": 0.85, "quality_score": 0.72},
    {"id": "SUP003", "name": "FastShip Co", "on_time_rate": 0.95, "quality_score": 0.90},
]
reasoner.load_ontology({s["id"]: s for s in suppliers})
print(f"✅ 加载了 {len(suppliers)} 个供应商")
```

## 4. 添加推理规则

```python
reasoner.add_rule({
    "id": "rule_delivery_risk",
    "condition": "on_time_rate < 0.85",
    "conclusion": "供应商准时率低于阈值，存在交付风险",
    "weight": 0.9
})
reasoner.add_rule({
    "id": "rule_quality_risk",
    "condition": "quality_score < 0.75",
    "conclusion": "供应商质量分数低于阈值，存在质量风险",
    "weight": 0.9
})
print("✅ 加载了2条推理规则")
```

## 5. 推理 + 置信度自知

```python
# 查询供应商的交付风险
result = reasoner.reason("delivery risk", context={"entity_id": "SUP002"})
print(f"供应商: SUP002 (Global Parts Ltd)")
print(f"推理结论: {result.conclusion}")
print(f"置信度: {result.confidence.value}")
print(f"使用的规则: {result.rules_used}")
```

```python
# Agent知道什么时候说"我不知道"
result_uncertain = reasoner.reason("market trend", context={"entity_id": "SUP001"})
if result_uncertain.confidence == ConfidenceLevel.SPECULATIVE:
    print("🤔 Agent: '这个问题超出了我的知识范围，我需要更多信息'")
```

## 6. 运行时学习

```python
# Agent在生产中学习新规则（不需要重新训练）
learner = AutoLearnEngine()
learner.learn(
    content="当 defect_rate > 0.08 且 supplier_tier < B 时，风险增加40%",
    confidence=0.94,
    source_type="expert_rule"
)
print("✅ Agent学会了新规则，无需重新训练！")
```
